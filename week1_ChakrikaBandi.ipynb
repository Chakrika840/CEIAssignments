**Your Geometric Explanation (4.3):**

Geometrically, a matrix can be viewed as a transformation that stretches, compresses, rotates, or shears vectors in space. Most vectors change both magnitude and direction after the transformation. However, certain special vectors, called eigenvectors, retain their direction and only get scaled. The scaling factor is called the eigenvalue. If the eigenvalue is greater than 1, the vector is stretched; if it lies between 0 and 1, the vector is compressed; and if it is negative, the vector is additionally flipped in direction. Thus, eigenvectors represent invariant directions of a transformation, while eigenvalues quantify how much stretching or compression occurs along those directions.

### 4.4 SVD & Dimensionality Reduction

1. Create a random 4×3 matrix `X` (use `np.random.seed(42)`).
2. Compute its **SVD**: `U, S, Vt = np.linalg.svd(X, full_matrices=False)`.
3. **Reconstruct** `X` from the SVD and verify it matches the original.
4. Perform a **rank-1 approximation** using only the largest singular value/vector — call it `X_approx`.
5. Explain the relationship between SVD components and PCA in the markdown cell below.

````python
np.random.seed(42)
X = np.random.randn(4, 3)

U, S, Vt = np.linalg.svd(X,full_matrices=False)

X_reconstructed = U@np.diag(S)@Vt
X_approx        = S[0]*np.outer(U[:,0],Vt[0,:])

print(f"U shape: {U.shape}, S shape: {S.shape}, Vt shape: {Vt.shape}")
print(f"Singular values: {S}")
print(f"Reconstruction error: {np.linalg.norm(X - X_reconstructed):.2e}")
print(f"Rank-1 approximation:\n{X_approx}")

assert np.allclose(X, X_reconstructed, atol=1e-10), "Reconstruction failed!"
print("4.4 passed ✓")
````

**SVD → PCA Connection (4.4):**
While studying PCA, I found that SVD provides an efficient way to compute the principal components of a dataset. For a data matrix X, SVD decomposes it into three matrices:

X = U * S * V^T

Initially, the three matrices looked confusing, but their roles become clearer when viewed geometrically. The matrix V contains the important directions present in the data. These directions are the same directions that PCA identifies as principal components. The matrix S contains singular values, which indicate how important each of these directions is. Larger singular values correspond to directions along which the data shows greater variation. The matrix U represents the data after it has been projected onto these new directions.

In PCA, our objective is to find the directions that capture the maximum variance in the dataset. When the data is centered, the columns of V obtained from SVD correspond directly to these principal component directions. The first column of V represents the direction with the highest variance, the second column represents the next most important direction, and so on.

Therefore, the matrix that contains the directions of maximum variance is V. The singular values in S tell us how much variance is captured along each of these directions. This connection between SVD and PCA makes dimensionality reduction computationally efficient while preserving most of the important information in the data.

### 5.4 Distribution Testing & Stationarity

1. Generate two samples: `s1` ~ Normal(0, 1), `s2` ~ Exponential(1), both n=200 (seed=0).
2. Run the **Kolmogorov-Smirnov test** (`stats.kstest`) comparing each against a standard normal CDF.
3. Generate a non-stationary time series (linear trend + noise). Run the **Augmented Dickey-Fuller test** (`adfuller` from `statsmodels`) and interpret the result.
4. Difference the series once and re-run ADF to confirm stationarity.

````python
from statsmodels.tsa.stattools import adfuller

np.random.seed(0)
s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1,'norm')
ks_s2 = stats.kstest(s2,'norm')
print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)
print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)
adf_diff = adfuller(ts_diff)
print(f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, Stationary? {adf_diff[1] < 0.05}")
````

### Distribution Testing and Stationarity Analysis

#### Kolmogorov-Smirnov (KS) Test

The KS test was used to compare two generated samples against a standard normal distribution.

* For **s1** (generated from a normal distribution), the p-value was **0.6268**, which is greater than 0.05. Therefore, we fail to reject the null hypothesis and conclude that the sample is consistent with a normal distribution.

* For **s2** (generated from an exponential distribution), the p-value was approximately **0.0000**, which is less than 0.05. Therefore, the null hypothesis is rejected, indicating that the sample does not follow a normal distribution.

#### Augmented Dickey-Fuller (ADF) Test

A non-stationary time series was generated using a linear trend combined with random noise. The ADF test was applied to determine whether the series was stationary.

* Before differencing, the ADF test produced a p-value of **0.7802**. Since the p-value is greater than 0.05, we fail to reject the null hypothesis of the ADF test, indicating that the series is non-stationary.

* The series was then differenced once to remove the trend component. After differencing, the ADF test returned a p-value close to **0.0000**. Since the p-value is less than 0.05, the null hypothesis is rejected, confirming that the differenced series is stationary.

#### Conclusion

The KS test successfully distinguished between normally distributed and non-normally distributed data. The ADF test showed that the original time series was non-stationary due to the presence of a trend, while first-order differencing transformed it into a stationary series suitable for time-series modeling.

````python
def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    eps = 1e-10
    breakpoints = np.linspace(min(expected.min(), actual.min()),
                               max(expected.max(), actual.max()), bins + 1)
    exp_counts, _ = np.histogram(expected, bins=breakpoints)
    act_counts, _ = np.histogram(actual,   bins=breakpoints)
    exp_pct = exp_counts / len(expected) + eps
    act_pct = act_counts / len(actual)   + eps
    return np.sum((act_pct - exp_pct) * np.log(act_pct / exp_pct))
np.random.seed(1)
train_dist = np.random.normal(50, 10, 1000)
drift_dist = np.random.normal(65, 12, 1000)
psi_value = compute_psi(train_dist, drift_dist)
print(f"PSI: {psi_value:.4f}")
print(f"Shift severity: {'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}")
fig, ax = plt.subplots()
ax.hist(train_dist, bins=30, density=True, alpha=0.5, color='steelblue', label='Training (original)')
ax.hist(drift_dist, bins=30, density=True, alpha=0.5, color='tomato',    label='Production (shifted)')
ax.legend(); ax.set_title('Distribution Shift Visualization')
plt.tight_layout(); plt.show()
````

### Concept Drift vs Covariate Drift & PSI Thresholds

* **Concept Drift:** Concept drift occurs when the relationship between the input features and the target variable changes over time. As a result, patterns learned by the model during training may no longer be valid, leading to reduced prediction accuracy.

* **Covariate Drift:** Covariate drift occurs when the distribution of the input features changes, while the relationship between the features and the target remains largely unchanged. This is commonly observed when production data differs from the data used during training.

* **PSI < 0.1:** Indicates little or no significant change in the distribution. The model is generally considered stable.

* **PSI 0.1–0.2:** Indicates a moderate shift in the distribution. The feature should be monitored closely as performance degradation may occur.

* **PSI > 0.2:** Indicates a significant distribution shift. This suggests that the model may be receiving data that differs substantially from the training data and further investigation is recommended.

* **Retraining Trigger Example:** A common retraining trigger is when one or more important features consistently show a PSI greater than 0.2 or when model performance metrics such as accuracy, precision, or F1-score fall below an acceptable threshold.

### 6.3 Bayes' Theorem

**Scenario — Spam filter:**
- P(Spam) = 0.30 (prior)
- P("free" | Spam) = 0.80 (likelihood)
- P("free" | Not Spam) = 0.05

1. Compute P(Spam | "free") using Bayes' theorem.
2. Implement a simple `naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham)` that returns the posterior P(Spam | word).
3. In the markdown cell, map each term — **prior, likelihood, evidence, posterior** — to the spam example.

````python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = p_free_given_spam*p_spam+p_free_given_ham*p_ham

# Bayes
p_spam_given_free = (p_free_given_spam*p_spam)/p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    prob=p_word_given_spam*prior_spam+p_word_given_ham*(1-prior_spam)
    return (p_word_given_spam*prior_spam)/prob
    pass

pred = naive_bayes_predict(0.30, 0.80, 0.05)
assert abs(pred - p_spam_given_free) < 1e-9
print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
````

#### Bayes Term Mapping (6.3)

| Term | Formula | In this example |
|------|---------|-----------------|
| Prior | P(Spam) | Probability that an email is spam before looking at any words. Here, P(Spam) = 0.30. |
| Likelihood | P("free" \| Spam) | Probability of seeing the word "free" given that the email is spam. Here, P("free" \| Spam) = 0.80. |
| Evidence | P("free") | Overall probability of seeing the word "free" in an email. Here, P("free") = 0.2750. |
| Posterior | P(Spam \| "free") | Probability that an email is spam after observing the word "free". Here, P(Spam \| "free") = 0.8727. |

**Interpretation:**  
Before seeing any words, there is a 30% chance that an email is spam. After observing the word **"free"**, the probability increases to about **87.27%**. This demonstrates how Bayes' theorem updates the probability of an event when new evidence becomes available.

### 6.4 Central Limit Theorem

1. Define a population following an **exponential distribution** (λ=1, size=100,000).
2. Repeatedly draw samples of size `n=30` and record their means — do this 5,000 times.
3. Plot the histogram of sample means.
4. Overlay the **theoretical normal distribution** predicted by the CLT: μ = population mean, σ = population_std / sqrt(n).
5. Run a KS test to confirm the distribution of sample means is approximately normal.

````python
np.random.seed(7)
population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30
sample_means = np.array([population[np.random.choice(len(population),sample_size,replace=False)].mean() for _ in range(n_samples)])
pop_mean = population.mean()
pop_std  = population.std()
clt_std  = pop_std/np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()
ax.hist(sample_means, bins=50, density=True, alpha=0.6, color='steelblue', edgecolor='white', label='Sample means')
x_clt = np.linspace(sample_means.min(), sample_means.max(), 300)
ax.plot(x_clt, stats.norm.pdf(x_clt, pop_mean, clt_std), color='navy', lw=2, label='Theoretical Normal (CLT)')
ax.legend()
ax.set_title('CLT: Distribution of Sample Means')
plt.tight_layout(); plt.show()
# KS test
ks_result = stats.kstest(sample_means, 'norm', args=(pop_mean, clt_std))
print(f"KS test p-value: {ks_result.pvalue:.4f} → Approximately normal? {ks_result.pvalue > 0.05}")
````

### CLT Reflection (6.4)

I found it interesting that although the population was exponentially distributed, the distribution of sample means was much closer to a normal distribution. This illustrates the Central Limit Theorem and explains why normal-distribution-based methods are widely used in practice.

The CLT provides the foundation for many statistical procedures, including confidence intervals and hypothesis tests, because it allows the sampling distribution of the mean to be treated as approximately normal.
